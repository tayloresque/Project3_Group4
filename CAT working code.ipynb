```python
import requests
import json
import pandas as pd
```

```python
url = "https://priceline-com-provider.p.rapidapi.com/v1/hotels/locations"

querystring = {"name":"SAN FRANCISCO","search_type":"HOTEL"}

headers = {
	"X-RapidAPI-Key": "3e6b744580msh4923cc9e6404566p18d83fjsnf66d6bfce325",
	"X-RapidAPI-Host": "priceline-com-provider.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)

print(response.json())

#if response.status_code == 200:
    # Use json.dumps with 'indent' parameter to pretty-print the JSON data
    #json_data = json.dumps(response.json(), indent=4, sort_keys=True)
    #print(json_data)
#else:
    #print("Failed to get data. Status code:", response.status_code)
```

```python

dictionary_data = json.loads(response.text)

SFhotel_data = pd.DataFrame(dictionary_data)

SFhotel_data.head()
```

```python
SFhotel_data.columns
```

```python
SFhotel_data['Hotel Name'] = SFhotel_data['itemName'].str.split('-').str[0]
SFhotel_data['City'] = SFhotel_data['itemName'].str.split('-').str[1]

column_to_select = ['Hotel Name', 'City', 'id', 'type',
                    'lat', 'lon','proximity', 'cityName', 
                    'score','radius', 'zip', 
                    ]

SFhotel_data_select = SFhotel_data[column_to_select]

SFhotel_data_select
```
