```python
# Dependencies and Setup
import requests
import json
import pandas as pd
import os
```

```python
city_list = [
    '3000002244',
    '',
    '3000002241',
    '3000001438',
    '3000002228',
    '3000002117',
    '3000002270',
    '3000002028',
    '3000001750',
    '3000001942',
    '3000002277',
    '3000002049',
    '3000002060',
    '3000002273',
    '3000002250',
]

# Initialize an empty list to store DataFrames
dfs = []

# Loop through each city in the list
for city in city_list:
    url = "https://priceline-com-provider.p.rapidapi.com/v1/hotels/search"
    
    querystring = {"date_checkout":"2023-10-19","sort_order":"PRICE","location_id": city_list,"date_checkin":"2023-10-18","amenities_ids":"FINTRNT,FBRKFST","rooms_number":"1","star_rating_ids":"3.0,3.5,4.0,4.5,5.0"}

    headers = {
        "X-RapidAPI-Key": "",
        "X-RapidAPI-Host": "priceline-com-provider.p.rapidapi.com"
    }
    
    response = requests.get(url, headers=headers, params=querystring)

    if response.status_code == 200:

        # Load the JSON response data into a list
        list_data = json.loads(response.text)
        
        # Create a DataFrame for the current city and append it to the list
        hotel_data = pd.DataFrame(list_data)
        dfs.append(hotel_data)

        
        # Get the JSON data
        json_data = response.json()
        
        # Use json.dumps with 'indent' parameter to pretty-print the JSON data
        json_data = json.dumps(response.json(), indent=4)
        
        print(f"Hotel data for {city}:")
        print(json_data)
    else:
        print(f"Failed to get data for {city}. Status code:", response.status_code)
```

```python

dictionary_data = json.loads(response.text)

hotel_data = pd.DataFrame(dictionary_data)

hotel_data.head()
```
